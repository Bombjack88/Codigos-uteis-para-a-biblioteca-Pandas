```python
# criar uma função
def tratamento(nome, sobrenome, ano_de, ano_ate):
    '''
    Primeiras 3 letras do nome
    Contar a quantidade de letras do sobrenome e manter a última metade,
    Calcular a idade até 2023
    '''
    nome=nome[:3]
    sobrenome_manter=int(round(len(sobrenome)/2,0))
    sobrenome=sobrenome[-sobrenome_manter:]
    idade=str(ano_ate-ano_de)
    txt=f'{nome}|{sobrenome}|{idade}'
    return txt
    
```
